## Making the Model

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import random
import csv
from tqdm import tqdm
```

```python
#load all data
cpu = pd.read_csv("new_cpu_with_features.csv")
gpu = pd.read_csv("new_gpu_with_features.csv")
ram = pd.read_csv("new_ram_with_features.csv")
psu = pd.read_csv("new_psu.csv")
storage = pd.read_csv("new_storage.csv")
motherboard = pd.read_csv("new_motherboard.csv")
case = pd.read_csv("new_case.csv")
```

```python
cpu = cpu.rename(columns={'price': 'cpu_price', 'Benchmark': 'cpu_benchmark', 'tdp': 'cpu_tdp', 'avg_p2p': 'cpu_avg_p2p'})
gpu = gpu.rename(columns={'price': 'gpu_price', 'Benchmark': 'gpu_benchmark', 'tdp': 'gpu_tdp', 'avg_p2p': 'gpu_avg_p2p'})
motherboard = motherboard.rename(columns={'price': 'motherboard_price'})
psu = psu.rename(columns={'price': 'psu_price', 'wattage': 'psu_wattage'})
ram = ram.rename(columns={'price': 'ram_price'})
storage = storage.rename(columns={'price': 'storage_price'})
case = case.rename(columns={'price': 'case_price', 'max_gpu_length': 'case_max_gpu_length'})

cpu['cpu_tdp'] = pd.to_numeric(cpu['cpu_tdp'], errors='coerce')
gpu['gpu_tdp'] = pd.to_numeric(gpu['gpu_tdp'], errors='coerce')
psu['psu_wattage'] = pd.to_numeric(psu['psu_wattage'], errors='coerce')
motherboard['motherboard_price'] = pd.to_numeric(motherboard['motherboard_price'], errors='coerce')
ram['ram_price'] = pd.to_numeric(ram['ram_price'], errors='coerce')
storage['storage_price'] = pd.to_numeric(storage['storage_price'], errors='coerce')
case['case_price'] = pd.to_numeric(case['case_price'], errors='coerce')
```

```python
def create_tiers(df, column, n_tiers=3):
    df['tier'] = pd.qcut(df[column], q=n_tiers, labels=['budget', 'mid', 'high'])
    return df

def generate_balanced_config(target_budget):
    cpu_tiered = create_tiers(cpu, 'cpu_price')
    gpu_tiered = create_tiers(gpu, 'gpu_price')
    psu_tiered = create_tiers(psu, 'psu_price')
    motherboard_tiered = create_tiers(motherboard, 'motherboard_price')
    ram_tiered = create_tiers(ram, 'ram_price')
    storage_tiered = create_tiers(storage, 'storage_price')
    case_tiered = create_tiers(case, 'case_price')

    if target_budget < 800:
        cpu_tier = gpu_tier = 'budget'
        other_tier = 'budget'
    elif 800 <= target_budget < 1500:
        cpu_tier = gpu_tier = 'mid'
        other_tier = 'budget'
    else:
        cpu_tier = gpu_tier = 'high'
        other_tier = 'mid'

    cpu_item = cpu_tiered[cpu_tiered['tier'] == cpu_tier].sample(1).iloc[0]
    gpu_item = gpu_tiered[gpu_tiered['tier'] == gpu_tier].sample(1).iloc[0]

    compatible_motherboards = motherboard_tiered[
        (motherboard_tiered['socket'] == cpu_item['socket']) & 
        (motherboard_tiered['tier'] == other_tier)
    ]
    motherboard_item = compatible_motherboards.sample(1).iloc[0] if not compatible_motherboards.empty else None

    total_power_consumption = cpu_item['cpu_tdp'] + gpu_item['gpu_tdp'] + 50
    compatible_psus = psu_tiered[
        (psu_tiered['psu_wattage'] > (total_power_consumption * 1.3)) & 
        (psu_tiered['tier'] == other_tier)
    ]
    psu_item = compatible_psus.sample(1).iloc[0] if not compatible_psus.empty else None

    compatible_cases = case_tiered[
        (case_tiered['case_max_gpu_length'] >= gpu_item['length']) & 
        (case_tiered['tier'] == other_tier)
    ]
    case_item = compatible_cases.sample(1).iloc[0] if not compatible_cases.empty else None

    ram_item = ram_tiered[
        (ram_tiered['total_ram'] >= 16) & 
        (ram_tiered['total_ram'] <= 32) & 
        (ram_tiered['tier'] == other_tier)
    ].sample(1).iloc[0]

    storage_item = storage_tiered[storage_tiered['tier'] == other_tier].sample(1).iloc[0]

    if motherboard_item is None or psu_item is None or case_item is None:
        return None

    return (cpu_item, gpu_item, psu_item, motherboard_item, ram_item, storage_item, case_item)

def evaluate_configuration(config):
    cpu, gpu, psu, motherboard, ram, storage, case = config
    total_cost = (cpu['cpu_price'] + gpu['gpu_price'] + psu['psu_price'] +
                  motherboard['motherboard_price'] + ram['ram_price'] +
                  storage['storage_price'] + case['case_price'])
    
    cpu_weight = 0.49
    gpu_weight = 0.49
    ram_weight = 0.01
    storage_weight = 0.01
    
    storage_speed_factor = {
        'M.2-2280': 1.0, 'PCIe': 1.0, 'M.2-2242': 0.9, 'M.2-22110': 1.0,
        'M.2-2230': 0.9, 'M.2-2260': 0.95, '2.5': 0.8, '3.5': 0.7
    }
    
    cpu_score = (cpu['cpu_benchmark'] * cpu['cpu_avg_p2p'])
    gpu_score = (gpu['gpu_benchmark'] * gpu['gpu_avg_p2p'])
    ram_score = (ram['total_ram'] * ram['speed'] * ram['speed_to_price']) 
    storage_score = (storage['capacity'] * storage_speed_factor.get(storage['type'], 0.8))

    performance_metric = (
        cpu_weight * cpu_score +
        gpu_weight * gpu_score +
        ram_weight * ram_score +
        storage_weight * storage_score
    )

    value_score = performance_metric / total_cost
    composite_score = (performance_metric * 0.7 + value_score * 0.3) * 100

    return {
        'cpu': f"{cpu['name']}",
        'cpu_brand' : f"{cpu['Brand']}",
        'gpu': f"{gpu['name']} ({gpu['Model']})",
        'gpu_brand': f"{gpu['Brand']}",
        'psu': f"{psu['name']} ({psu['psu_wattage']}W)",
        'motherboard': motherboard['name'],
        'ram': f"{ram['name']} ({ram['total_ram']}GB, {ram['speed']}MHz)",
        'storage': f"{storage['name']} ({storage['capacity']} {storage['type']})",
        'case': case['name'],
        'cpu_benchmark': cpu['cpu_benchmark'],
        'gpu_benchmark': gpu['gpu_benchmark'],
        'cpu_price': cpu['cpu_price'],
        'gpu_price': gpu['gpu_price'],
        'psu_price': psu['psu_price'],
        'motherboard_price': motherboard['motherboard_price'],
        'ram_price': ram['ram_price'],
        'storage_price': storage['storage_price'],
        'case_price': case['case_price'],
        'total_cost': total_cost,
        'performance_metric': performance_metric,
        'value_score': value_score,
        'composite_score': composite_score
    }
def generate_dataset(initial_samples=100000, batch_size=100000, max_samples=1000000, output_file='configurations.csv'):
    fieldnames = ['cpu', 'cpu_brand', 'gpu', 'gpu_brand', 'psu', 'motherboard', 'ram', 'storage', 'case',
                  'cpu_benchmark', 'gpu_benchmark', 'cpu_price', 'gpu_price', 'psu_price', 'motherboard_price',
                  'ram_price', 'storage_price', 'case_price', 'total_cost',
                  'performance_metric', 'value_score', 'composite_score']
    
    configs = []
    total_samples = 0
    
    with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        
        print(f"Generating initial {initial_samples} samples...")
        for _ in tqdm(range(initial_samples)):
            target_budget = random.uniform(500, 2500)
            config = generate_balanced_config(target_budget)
            if config:
                evaluated_config = evaluate_configuration(config)
                writer.writerow(evaluated_config)
                configs.append(evaluated_config)
                total_samples += 1
        
        while total_samples < max_samples:
            print(f"\nAnalyzing current dataset ({total_samples} samples)...")
            df = pd.DataFrame(configs)
            
            budget_distribution = pd.cut(df['total_cost'], bins=5).value_counts(normalize=True)
            performance_distribution = pd.cut(df['performance_metric'], bins=5).value_counts(normalize=True)
            print("Budget distribution:")
            print(budget_distribution)
            print("\nPerformance distribution:")
            print(performance_distribution)
            
            if budget_distribution.index[0] > budget_distribution.index[-1]:
                budget_range = (500, 1500) 
            else:
                budget_range = (1000, 3000) 
            
            print(f"\nGenerating additional {batch_size} samples...")
            new_configs = []
            for _ in tqdm(range(batch_size)):
                target_budget = random.uniform(*budget_range)
                config = generate_balanced_config(target_budget)
                if config:
                    evaluated_config = evaluate_configuration(config)
                    writer.writerow(evaluated_config)
                    new_configs.append(evaluated_config)
                    total_samples += 1
                
                if total_samples >= max_samples:
                    break
            
            configs.extend(new_configs)
            
            print(f"\nTotal samples generated: {total_samples}")
            
            #user_input = input("Generate another batch? (y/n): ").lower()
            #if user_input != 'y':
                #break
    
    print(f"\nDataset generation complete. Total samples: {total_samples}")
    print(f"Dataset saved to {output_file}")

generate_dataset()
```

## Creating the KMeans model to assign clusters and provide diverse configurations

```python
df = pd.read_csv('configurations.csv')
features = ['cpu_benchmark', 'gpu_benchmark', 'cpu_price', 'gpu_price', 'psu_price', 
            'motherboard_price', 'ram_price', 'storage_price', 'case_price','performance_metric', 'value_score']
scaler = StandardScaler()
X = scaler.fit_transform(df[features])
kmeans = KMeans(n_clusters=25)
df['cluster'] = kmeans.fit_predict(X)
df.to_csv("clustered_configs.csv", index=False)
df
```

```python
def recommend_build(budget, cpu_pref=None, gpu_pref=None):
    df = pd.read_csv("clustered_configs.csv")
    
    filtered_df = df[(df['total_cost'].between(budget*0.9, budget*1.1))]
    if cpu_pref and gpu_pref:
        filtered_df = filtered_df[(df['cpu_brand'].str.contains(cpu_pref)) & 
                         (df['gpu_brand'].str.contains(gpu_pref))]
    elif cpu_pref and not gpu_pref:
        filtered_df = filtered_df[(df['cpu_brand'].str.contains(cpu_pref))]
    elif gpu_pref and not cpu_pref:
        filtered_df = filtered_df[(df['gpu_brand'].str.contains(gpu_pref))]
    
    if(len(filtered_df) == 0):
        return "No compatible matches with the budget and preferences found!"
    
    recommendations = filtered_df.loc[filtered_df.groupby('cluster')['composite_score'].idxmax()]
    recommendations.performance_metric = recommendations.performance_metric / 1000
    recommendations.value_score = recommendations.value_score / 1000
    recommendations.composite_score = recommendations.composite_score / 10000
    return recommendations.sort_values(by=['cpu_benchmark', 'gpu_benchmark', 'composite_score'], ascending=False).reset_index().drop(['index'], axis=1).head(5)
```
